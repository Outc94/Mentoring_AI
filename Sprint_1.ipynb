## Problem 1

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('application_train.csv')
```

```python
df2 = df
list_str=[]
for i in df.columns:
    if df[i].dtypes == 'O':
        df2 = df2.drop([i], axis = 1)
        list_str.append(i)
df2 = df2.replace(np.nan,0)
df_str = df[list_str]
df_dummy = pd.get_dummies(df_str)
df_mix = pd.concat([df2, df_dummy], axis=1)
```

```python
df_dummy.head(10)
```

```python
df_mix.head(10)
```

```python
from scipy.stats import pearsonr
list_i = []
for i in df_mix.columns:
    a, b = pearsonr(df_mix[i],df_mix["TARGET"])
    if 0.06 < abs(a) :
        list_i.append(i)
        
list_i = list_i[1:]
X = df_mix[list_i]
y = df_mix["TARGET"]
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_std = sc.fit_transform(X)
```

Now we will use the KFold class to devide it multiple times

```python
from sklearn.model_selection import KFold
kfold = KFold(n_splits=10)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

reg = KNeighborsClassifier(n_neighbors=3)
cross_val_score(reg, X, y, cv=kfold)
```

```python
cross_val_score(reg, X, y, cv=kfold).mean()
```

```python
(X_train, X_test,
 y_train, y_test) = train_test_split(
    X, y, test_size=0.25, random_state=0,
)
reg = KNeighborsClassifier(n_neighbors=3)
reg.fit(X_train, y_train)
reg_pred = reg.predict(X_test)
```

```python
df2 = df
list_str=[]
for i in df.columns:
    if df[i].dtypes == 'O':
        df2 = df2.drop([i], axis = 1)
        list_str.append(i)
df2 = df2.replace(np.nan,0)
df_str = df[list_str]
df_dummy = pd.get_dummies(df_str)
df_mix = pd.concat([df2, df_dummy], axis=1)
```

```python
df_mix["NA_count"] = pd.DataFrame(df.isnull().sum(axis=1))
df_mix.head()
```

```python
list_i = []
for i in df_mix.columns:
    a, b = pearsonr(df_mix[i],df_mix["TARGET"])
    if 0.04 < abs(a) :
        list_i.append(i)
        
X = df_mix[list_i].drop("TARGET", axis=1)
y = df_mix["TARGET"]

(X_train, X_test,
 y_train, y_test) = train_test_split(
    X, y, test_size=0.25, random_state=0,
)
```

Let's see the number of outliers, if the number is 1 then we have a normal situation.

```python
from sklearn.neighbors import LocalOutlierFactor
clf = LocalOutlierFactor(n_neighbors=10)
y_pred = clf.fit_predict(X_train)
ANOMALY_DATA = -1
predicted_outlier_index = np.where(y_pred == ANOMALY_DATA)
print("Number of outliers: {}".format(len(predicted_outlier_index)))
X_train = X_train.iloc[predicted_outlier_index]
y_train = y_train.iloc[predicted_outlier_index]
```

```python
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
y_train = np.array(y_train)
y_test = np.array(y_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.model_selection import GridSearchCV
```

```python
clf = lgb.LGBMClassifier().fit(X_train, y_train)
y_pred =clf.predict(X_test)
print(clf)
print("{}".format(clf.score(X_test, y_test)))
print("{}".format(precision_score(y_test,y_pred)))
print("{}".format(recall_score(y_test,y_pred)))
print("{}".format(f1_score(y_test,y_pred)))
print(confusion_matrix(y_test,y_pred))
```

```python
top_features = clf.feature_importances_.reshape(len(clf.feature_importances_),1)
column = np.array(X.columns).reshape(len(clf.feature_importances_),1)
feature = pd.DataFrame(np.hstack([column,top_features]), columns=["Columns", "Top Features"])
feature_sorted = feature.sort_values('Top Features', ascending=False).reset_index()
feature_sorted = feature_sorted.drop("index", axis=1)
feature_sorted.head()
```

```python
feature_sorted["Top Features"] = feature_sorted["Top Features"].astype('float64')
list_col = []
for i in range(len(feature_sorted)):
    a = feature_sorted.iloc[i,0]
    b = feature_sorted.iloc[i,1]
    if b >= 30:
        list_col.append(a)
```

```python
list_col
```

```python
X = X[list_col]
(X_train, X_test,
 y_train, y_test) = train_test_split(
    X, y, test_size=0.25, random_state=0,
)
sc = StandardScaler()
X_train_std = sc.fit_transform(X_train)
X_test_std = sc.transform(X_test)
```

```python
clf = lgb.LGBMClassifier(max_depth=5, n_estimators=100).fit(X_train, y_train)
y_pred =clf.predict(X_test)
print(clf)
print("{}".format(clf.score(X_test, y_test)))
print("{}".format(precision_score(y_test,y_pred)))
print("{}".format(recall_score(y_test,y_pred)))
print("{}".format(f1_score(y_test,y_pred)))
print(confusion_matrix(y_test,y_pred))
```

```python
clf = lgb.LGBMClassifier(max_depth=3, n_estimators=1000).fit(X_train, y_train)
y_pred =clf.predict(X_test)
print(clf)
print("{}".format(clf.score(X_test, y_test)))
print("{}".format(precision_score(y_test,y_pred)))
print("{}".format(recall_score(y_test,y_pred)))
print("{}".format(f1_score(y_test,y_pred)))
print(confusion_matrix(y_test,y_pred))
```
